EXPLORING EBAY CAR SALES

The data set is of used cars from ebay Kleinanzeigen gotten from dataquest. The columns of the dataset are 20 and are described as follows:


dateCrawled - When this ad was first crawled. All field-values are taken from this date.
name - Name of the car.
seller - Whether the seller is private or a dealer.
offerType - The type of listing
price - The price on the ad to sell the car.
abtest - Whether the listing is included in an A/B test.
vehicleType - The vehicle Type.
yearOfRegistration - The year in which the car was first registered.
gearbox - The transmission type.
powerPS - The power of the car in PS.
model - The car model name.
kilometer - How many kilometers the car has driven.
monthOfRegistration - The month in which the car was first registered.
fuelType - What type of fuel the car uses.
brand - The brand of the car.
notRepairedDamage - If the car has a damage which is not yet repaired.
dateCreated - The date on which the eBay listing was created.
nrOfPictures - The number of pictures in the ad.
postalCode - The postal code for the location of the vehicle.
lastSeenOnline - When the crawler saw this ad last online.


OBJECTIVE

The aim is to clean the data and and do a basic analysis to find out some interesting trends such as most expensive car brands, top selling brands and attempt to find out if theres is some correlation between variables.

REMOVING NULL VALUES

We see that certain columns contain null values for which price and odometer have corresponding values. We will remove these rows with null values for the 'price' and 'odometer_km' columns while also removing the 'unrepaired_damage' column as there is an abundance of null values and the column is of little interest or no significance.

```python
autos_19_2016_cleaner = autos_19_2016.loc[(autos_19_2016['price'].notnull()) & (autos_19_2016['odometer_km'].notnull())]
autos_19_2016_cleaner.drop(['unrepaired_damage'], axis = 1, inplace = True)
autos_19_2016_cleaner.info()
```

ANALYSIS 

We want to aggregate on the brand column by identifying unique values and counting the unique values.

```python
#creating an empty dictionary
#autos_19_2016_cleaner['brand'].value_counts().sort_index
#brand_dict = {}
autos_19_2016_cleaner['brand'].nunique()
    
```

Estimating top 20 selling brands

```python
auto_brands_20 = autos_19_2016_cleaner['brand'].value_counts().head(20).index
auto_brands_20
```

Estimating the mean selling price for the top 20 selling car brands that will be saved to a dictionary

```python
#create dictionary
brand_mean_prices = {}
for brand in auto_brands_20:
    mean_price = round(autos_19_2016_cleaner.loc[autos_19_2016_cleaner['brand'] == brand, 'price'].mean())
    # Creating a new key_value pair with this mean price
    brand_mean_prices[brand] = mean_price
```

```python
brand_mean_prices
```

```python
# sorting the dictionary key values 
brand_mean_prices = dict(sorted(brand_mean_prices.items(), key = lambda data: data[1], reverse = True ))
brand_mean_prices
```

From sorting the dictionary values, in descending order, the highest selling prices among top 20 was mini, audi and mercedes while renault, fiat and opel were least expenses among the top 20. The price gap between the highest and lowest selling is almost 2.5 times. Mazda and Citreon are mid range priced cars.

Aggregating top 6 brands using the Series constructor method

```python
brand_mean_prices = {}
brand_mean_mileage = {}
for brand in auto_brands_20:
    mean_price = round(autos_19_2016_cleaner.loc[autos_19_2016_cleaner['brand'] == brand, 'price'].mean())
    mean_mileage = round(autos_19_2016_cleaner.loc[autos_19_2016_cleaner['brand'] == brand, 'odometer_km'].mean())
    # Creating a new key_value pair with this mean price and mean mileage
    brand_mean_prices[brand] = mean_price
    brand_mean_mileage[brand] = mean_mileage
```

```python
brand_mean_prices
```

```python
brand_mean_mileage
```

```python
# sorting the values 
brand_mean_prices = dict(sorted(brand_mean_prices.items(), key = lambda data: data[1], reverse = True ))
brand_mean_mileage = dict(sorted(brand_mean_mileage.items(), key = lambda data: data[1], reverse = True ))
```

```python
brand_mean_prices
```

```python
brand_mean_mileage
```

The benz audi and opel seem to have the most mileage of the cars while the mini, smart and sonstige autos have the lowest mileage of the top 20 brands. We use the Series and Df constrcutor to aggragate these two column variables.
We convert both dictioanries to Series objects (mean price & mileage dictionary)

```python
mp_series = pd.Series(brand_mean_prices)
mileage_series = pd.Series(brand_mean_mileage)

print(mp_series)
print(mileage_series)
```

```python
# creating a dataframe from the first series (mean_price)
brands_top20 = pd.DataFrame(mp_series, columns=['mean_price'])
brands_top20
```

```python
# assign the other series(mileage) as a new column in the dataframe above

#brands_top20.assign(column=mileage_series)

brands_top20['mean_mileage_km'] =  mileage_series 
```

```python
# sorting the mean mileage values, since the one for price has been sorted akready
brands_top20.sort_values('mean_mileage_km', ascending = False)
```

It can be seen that the cars with lower mileage sell higher than cars with higher mileage. looking at the exploration below, it is seen that the Mini brand has the max selling price as well as the lowest mileage. 

```python
# Exploring the data simoltaneously to check if any relationship between 'price' and 'mileage' if any

print(brands_top20['mean_price'].idxmax())
print(brands_top20['mean_price'].idxmin())
```

```python
# Exploring the data simoltaneously to check if any relationship between 'price' and 'mileage'
print(brands_top20['mean_mileage_km'].idxmax())
print(brands_top20['mean_mileage_km'].idxmin())
```
